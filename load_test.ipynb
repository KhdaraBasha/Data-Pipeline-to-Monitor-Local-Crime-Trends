```python
# Import necessary modules
from sqlalchemy import create_engine
import psycopg2
from dotenv import load_dotenv
import os
```

```python
# Load PostgreSQL Connection details from the .config file
load_dotenv()

host = os.getenv('DB_HOST')
database = os.getenv('DB_NAME')
user = os.getenv('DB_USER')
password = os.getenv('DB_PASS')
port = os.getenv('DB_PORT')
```

```python
# Create PostgreSQL connection id
def fn_db_connection():
    '''
    '''

    load_dotenv()

    # data base connection credentials
    host = os.getenv('DB_HOST')
    database = os.getenv('DB_NAME')
    user = os.getenv('DB_USER')
    password = os.getenv('DB_PASS')
    port = os.getenv('DB_PORT')
    try:
        conn = psycopg2.connect\
                    (
                        host=host,
                        database=database,
                        user=user,
                        password=password,
                        port=port
                    )
        
        return conn
    
    except Exception as e:
        print(f'Error While Creating Dtabase Connection {e}') 
```

```python
# Create Table in The PostgreSQL
def fn_create_table():
    '''
    '''
    try:
        # create cursor object to execute SQL
        conn = fn_db_connection()
        cur = conn.cursor()

        # Create Table SQL Query
        create_table_query = '''
            CREATE TABLE IF NOT EXISTS cpd_db.tb_cpd_incidents (
                date_time TIMESTAMP,
                id INTEGER PRIMARY KEY,
                type TEXT,
                subtype TEXT,
                location TEXT,
                description TEXT,
                last_updated TIMESTAMP,
                year INTEGER,
                month INTEGER,
                day INTEGER,
                hour INTEGER, 
                minute INTEGER,
                second INTEGER
            )
        '''

        # Execute query to create table
        cur.execute(create_table_query)

        # Commit changes
        conn.commit()
    except Exception as e:
        print(f'Error while creating table {e}')
    finally:
        # close the cursor and connection
        cur.close()
        conn.close()
```

```python
# Load data into postgresql table
def fn_load_data_table(df):
    '''
    Loads the transformed data passed in as a DataFrame 
    into the 'cpd_incidents' table in our Postgres instance.
    '''
    # data base connection credentials
    host = os.getenv('DB_HOST')
    database = os.getenv('DB_NAME')
    user = os.getenv('DB_USER')
    password = os.getenv('DB_PASS')
    port = os.getenv('DB_PORT')

    # Create table to insert data into as necessary
    fn_create_table()

    # 
    engine = create_engine(f"postgresql://{user}:{password}@{host}:{port}/{database}")

    # insert data into Postgres DB into the 'cpd_incidents' table
    df.to_sql('cpd_db.tb_cpd_incidents', engine, if_exists='replace')
```
